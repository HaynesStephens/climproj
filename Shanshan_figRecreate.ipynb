### EQ FIGURE 4

```python
def Shanshan_EQ_fig4(ccsm3_Tsurf, ccsm3_netLWsurf, climt_Tsurf, climt_netLWsurf):
    fig, ax = plt.subplots(figsize=(10,10))
    ax.scatter(ccsm3_Tsurf, ccsm3_netLWsurf, c='k', s=17, label = 'CCMS3')
    ax.plot(climt_Tsurf, climt_netLWsurf, marker='o', c='r', markersize=10, label = 'CliMT')
    ax.set_ylabel('Net surface longwave (Wm$^{-2}$)')
    ax.set_xlabel('Surface temperature (K)')
    ax.legend()
```

```python
def Shanshan_EQ_fig5(vary_sol_dT, vary_sol_dF, vary_co2_dT, vary_co2_dF, vary_co2_cst_q_dT, vary_co2_cst_q_dF):
    fig, axes = plt.subplots(1, 3, figsize=(10,3))
    title_size = 9
    
    ax0 = axes[0]
    ax0.set_title('Varying solar', fontsize=title_size)
    ax0.set_ylabel('Wm$^{-2}$')
    ax0.set_xlabel('DT (K)')
    
    ax1 = axes[1]
    ax1.set_title('Varying CO$_2$', fontsize=title_size)
    ax1.set_xlabel('DT (K)')
    
    ax2 = axes[2]
    ax2.set_title('Varying CO$_2$: constant humidity', fontsize=title_size)
    ax2.set_xlabel('DT (K)')
```

### EQ FIGURE 6

```python
def Shanshan_EQ_fig6(climt_dT, climt_dLH, climt_q_dT, climt_q_dLH):
    fig, ax = plt.subplots(figsize=(10,10))
    ax.plot(climt_dT, climt_dLH, marker='o', c='k', markersize=7, label = 'CliMT')
    ax.plot(climt_q_dT, climt_q_dLH, marker='o', c='r', markersize=7, label = 'CliMT')
    ax.set_ylabel('DLH (Wm$^{-2}$)')
    ax.set_xlabel('DT (K)')
    ax.axhline(0, linestyle='--', linewidth=1, c='k')
    
    ax_twin = ax.twinx()
    ax_twin.set_ylabel('DLH cm/yr)')
    
    ax.legend()
```

```python
Shanshan_EQ_fig6([], [], [], [])
```

### EQ FIGURE 8

```python
########################33
# USE GRID LAYOUT FOR PLOTS W/ PADDING FOR FONT AND TITLE
########################33


def Shanshan_EQ_fig8():
    fig, axes = plt.subplots(2, 3, figsize=(10,7))
    label_size = 8
    title_size = 10
    
    ax0 = axes[0,0]
    ax0.set_title('Latent heat', fontsize=title_size)
    ax0.set_ylabel('Wm$^{-2}$', fontsize=label_size)
    
    ax1 = axes[0,1]
    ax1.set_title('LW allsky', fontsize=title_size)

    ax2 = axes[0,2]
    ax2.set_title('LW clearsky', fontsize=title_size)
    
    ax3 = axes[1,0]
    ax3.set_title('Sensible heat', fontsize=title_size)
    ax3.set_ylabel('Wm$^{-2}$', fontsize=label_size)
    ax3.set_xlabel('DT (K)', fontsize=label_size)
    
    ax4 = axes[1,1]
    ax4.set_title('SW allsky', fontsize=title_size)
    ax4.set_xlabel('DT (K)', fontsize=label_size)
    
    ax5 = axes[1,2]
    ax5.set_title('SW clearsky', fontsize=title_size)
    ax5.set_xlabel('DT (K)', fontsize=label_size)
    
    plt.tight_layout()
    fig.suptitle('Surface radiative anomalies in equilibrium state', fontsize=16)
```

```python
Shanshan_EQ_fig8()
```
