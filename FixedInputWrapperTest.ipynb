```python
import matplotlib.pyplot as plt
import copy
from FixedInputWrapper import FixedInputWrapper
from FixedInputWrapper_v2 import FixedInputWrapper_v2
from sympl import (DataArray, AdamsBashforth, get_constant, set_constant, NetCDFMonitor)
import numpy as np
from datetime import timedelta
from climt import (EmanuelConvection, DryConvectiveAdjustment, RRTMGShortwave, RRTMGLongwave,
                   SlabSurface, SimplePhysics, get_default_state)
import pickle
```

```python
restart_file_name = '/Users/haynesstephens1/uchi/research/climproj/climt_files/control_fullstore/' \
                    'i270_320solar_fullstore/i270_320solar_fullstore.eq.pkl'
restart_file = open(restart_file_name, 'rb')
restart_state = pickle.load(restart_file)
control_q = restart_state['specific_humidity'].copy()
```

```python
radiation_sw = RRTMGShortwave(ignore_day_of_year=True)
radiation_lw = RRTMGLongwave()
state = get_default_state([radiation_lw, radiation_sw])
fixed_state_ref = {'specific_humidity': copy.deepcopy(state)['specific_humidity']}
fixed_state_ref['specific_humidity'].values[:] = control_q
print('FIXED STATE REF KEYS:', fixed_state_ref.keys())
```

```python
def WrapperCheck1(wrapper):
    q0 = np.zeros((28, 1, 1))
    q1 = np.zeros((28, 1, 1)) + 0.005
    q2 = np.reshape(np.linspace(0,0.1,28), (28, 1, 1))
    q_list = [q0, q1, q2]
    fig, axis = plt.subplots(3,3)
    for i in range(len(q_list)):
        q = q_list[i]
        radiation_sw = RRTMGShortwave(ignore_day_of_year=True)
        radiation_lw = RRTMGLongwave()
        state = get_default_state([radiation_lw, radiation_sw])
        state['specific_humidity'].values[:] = q
        fixed_state = {'specific_humidity': copy.deepcopy(state)['specific_humidity']}
        fixed_state['specific_humidity'].values[:] = control_q
        print('FIXED STATE KEYS:', fixed_state.keys())
        radiation_lw_fixed = wrapper(radiation_lw, fixed_state)
        radiation_sw_fixed = wrapper(radiation_sw, fixed_state)
        air_pressure = np.copy(state['air_pressure']).flatten()[::-1]
        axis[i, 0].plot(np.copy(state['specific_humidity']).flatten(), air_pressure)
        axis[i, 0].set_yscale('log')
        cooling = radiation_lw_fixed(state)[1]['air_temperature_tendency_from_longwave']
        axis[i, 1].plot(np.copy(state['specific_humidity']).flatten(), air_pressure)
        axis[i, 1].set_yscale('log')
        axis[i, 2].plot(np.copy(cooling).flatten(), air_pressure)
        axis[i, 2].set_yscale('log')
    axis[0,0].set_title('Starting Profile')
    axis[0,1].set_title('Returned Profile')
    axis[0,2].set_title('LW Tendency')
    plt.tight_layout()
```

```python
WrapperCheck1(FixedInputWrapper)
```

```python
WrapperCheck1(FixedInputWrapper_v2)
```
